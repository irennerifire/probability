__Задача 1__

О непрерывной равномерно распределённой случайной величине `B` известно, что её дисперсия равна 0.2. Можно ли найти правую границу величины `B` и её среднее значение зная, что левая граница равна 0.5? Если да, найдите их.

В данном случае пусть а = 0.5 - левая граница. Дисперсия D(B) = 0.2.  Так как распределение равномерное и непрерывное, то D(B) = (b-a)^2/12, где b - правая граница. Математическое ожидание будет равно M(B) = (a+b)/2, что, фактически, и является средним значением.

Найдем b. Решим уравнение (b-0.5)^2/12=0.2

(b-0.5)^2 = 2.4

|b-0.5| = sqrt(2.4)

|b-0.5| = 1.549

b = 2.049 или b = -1.049

Так как нам нужна правая граница, то выберем b = 2.049

```python
np.sqrt(2.4)
```

M(B) = (a+b)/2 = (0.5 + 2.049)/2 = 1.2745 - среднее значение.
